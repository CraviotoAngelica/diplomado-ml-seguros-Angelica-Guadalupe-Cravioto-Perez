# Ejercicio en clase sesion 2

```python
from datetime import date

# Datos
nombres = ['Ana Torres', 'Luis Pérez', 'María García', 'Carlos Ruiz']
edades = [28, 45, 32, 62]
emision = [date(2026,1,15), date(2026,2,1), date(2026,3,10), date(2026,4,1)]
prima_anual = [2400, 6200, 1900, 14500]
activa = [True, True, False, True]

hoy = date.today()
```

### 1.Imprime el nombre, edad y prima de la póliza índice 2 (la tercera). ¿Está activa?

```python
i = 2
print("--POLIZA 3--")
print(f"Nombre: {nombres[i]}")
print(f"Edad: {edades[i]} años")
print(f"Prima: $", round(prima_anual[i], 2))
print(f"Activa:", activa[i])
```

#### 2.Calcula cuántos días lleva vigente cada póliza desde su emisión hasta hoy. (hoy - emision[i]).days

```python
print("\n---DIAS DE VIGENCIA---")
for i in range(len(nombres)):
    dias = ( hoy - emision[i]).days
    print(f"{nombres[i]}: {dias} días")
```

#### 3.Calcula la prima mensual de cada asegurado (prima_anual / 12). Formatea con 2 decimales.

```python
print("\n--PRIMA MENSUAL--")
for i in range(len(nombres)):
    prima_mensual = prima_anual[i] / 12
    print(f"{nombres[i]}: ${round(prima_mensual, 2)} al mes")
```

#### 4.¿Cuántas pólizas están activas? (cuenta los True en la lista activa).

```python
activas = 0
for a in activa:
    if a == True:
        activas = activas + 1
        
print(f"\nNúmero de pólizas activas: {activas}")
```

#### 5.Bonus: ¿Cuál es la prima promedio de las pólizas activas? (suma primas activas / total activas)

```python
suma = 0

for i in range(len(nombres)):
    if activa[i]:
        suma += prima_anual[i]

promedio = suma / activas
print(f'Prima promedio activas: ${promedio:.2f}')
```
